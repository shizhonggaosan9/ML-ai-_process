## sklearn knn 

```python
from sklearn.model_selection import train_test_split #分割数据集为训练测试集
from sklearn.neighbors import KNeighborsClassifier # knn
from sklearn.metrics import confusion_matrix        # 制作混淆矩阵
from sklearn.metrics import classification_report   # ^^^^
```

### 读取数据集并分割X和Y，训练集和测试集

```python
# 分割X和Y
X,Y = file2matrix('./data/datingTestSet2.txt')

# XY分别分割数据集和训练集
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 0)
X_train, len(X_train),  len(X_test), len(Y_train)
```

### 训练数据集

```python
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier.fit(X_train, Y_train)
```

### 测试数据集并统计错误率

```python
# 测试得到的结果集
y_pred = classifier.predict(X_test)
print(y_pred)

# 错误率0.184
t = y_pred==Y_test
sum([x==False for x in t]) / len(Y_test)
```

### 制作混淆矩阵

```python
cm = confusion_matrix(Y_test, y_pred)
print(cm)
print(classification_report(Y_test, y_pred))
```
